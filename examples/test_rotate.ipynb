```python
import folium

print(folium.__version__)
```

```python
m = folium.Map(location=[41, -71], zoom_start=4)

icon = folium.Icon(prefix='fa', color='green', icon='arrow-up', angle=180)
folium.Marker([41, -72], popup='something something...', icon=icon).add_to(m)

icon = folium.Icon(prefix='fa', color='green', icon='arrow-up', angle=45)
folium.Marker([41, -75], popup='something something...', icon=icon).add_to(m)

icon = folium.Icon(prefix='fa', color='green', icon='arrow-up', angle=90)
folium.Marker([41, -78], popup='something something...', icon=icon).add_to(m)

m
```
