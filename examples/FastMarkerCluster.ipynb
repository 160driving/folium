```python
import numpy as np
import os

import folium
from folium import plugins

n = 1000

# Create a list of lists that correspond to [[lat, lon], [lat, lon]]
data = np.array([
    np.random.uniform(low=35, high=60, size=n),  # Random latitude
    np.random.uniform(low=-12, high=30, size=n) # Random longitude
]).T.tolist()

m = folium.Map([45., 3.], zoom_start=6)
mc = plugins.FastMarkerCluster(data)

m.add_children(mc)
m.save(os.path.join('results', 'FastMarkerCluster_simple.html'))

m
```

```python
# A custom callback can be passed to edit the markers. This should be a string representation of a valid JavaScript function.
callback = ('function (row) {' +
                     'var icon, marker;' +
                     'icon = L.AwesomeMarkers.icon({' +
                     'icon: "map-marker", markerColor: "red"});' + 
                     'marker = L.marker(new L.LatLng(row[0], ' +
                     'row[1]));' +
                     'marker.setIcon(icon);' +
                     'return marker;' +
                     '};')

m = folium.Map([45., 3.], zoom_start=6)
mc = plugins.FastMarkerCluster(data, callback=callback)

m.add_children(mc)
m.save(os.path.join('results', 'FastMarkerCluster_custom.html'))

m
```
