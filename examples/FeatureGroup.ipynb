```python
from folium import Map, FeatureGroup, Marker, LayerControl

mapa = Map(location=[45.372, -121.6972], zoom_start=12, tiles='Stamen Terrain')

feature_group = FeatureGroup(name='Some icons')
Marker([45.3288, -121.6625], popup='Mt. Hood Meadows').add_to(feature_group)
Marker([45.3311, -121.7113], popup='Timberline Lodge').add_to(feature_group)

feature_group.add_to(mapa)
LayerControl().add_to(mapa)

mapa
```
